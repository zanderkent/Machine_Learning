```python
#Import required libraries
import pandas as pd
import numpy as np
#Import libraries:
import pandas as pd
import numpy as np
import xgboost as xgb
from xgboost.sklearn import XGBClassifier
from sklearn.model_selection import cross_validate
from sklearn.model_selection import train_test_split
from sklearn.model_selection import RandomizedSearchCV
```

```python
#Import data saved from data cleaning
X_train = pd.read_csv('training_features.csv', sep='\t')
Y_train = pd.read_csv('training_target.csv', sep='\t')
#Observe that is a class imbalance 80-20, should be fine. 
Y_train['y'].value_counts(normalize=True)
X_train, X_test, y_train, y_test = train_test_split(np.array(X_train), Y_train)
```

```python
#There was some swapping of values in here, 
#such as decreasing the learning rate, 
#increasing the number of estimators
#There was not much tuning that was required, since I reached approx 99% accuarcy quickly. 
#If needed I could do a gridsearch or a randomgrid search to improve the model.
model = XGBClassifier(
 learning_rate =0.01,
 n_estimators=4000,
 max_depth=5,
 min_child_weight=1,
 gamma=0,
 subsample=0.8,
 colsample_bytree=0.8,
 objective= 'binary:logistic',
 scale_pos_weight=1,
 n_jobs = -1,tree_method='gpu_hist',
 silent=False,verbose_eval=10,
 seed=27)
```

```python
#Cross validation in order to avoid over-fitting and ensure best parameters
#xgbtrain = xgb.DMatrix(np.array(X_train), label=np.array(Y_train))
#cvresult = xgb.cv(xgb1.get_xgb_params(), xgbtrain, num_boost_round=5000, nfold=5,
#            metrics='auc', early_stopping_rounds=10, verbose_eval=100)
```

```python
model.fit(X_train,y_train,eval_set=[(X_train, y_train), (X_test, y_test)],eval_metric='auc')
```

```python
#Check accuracy
X_test = pd.read_csv('testing_targets.csv', sep='\t')
preds = model.predict_proba(np.array(X_test))
pd.DataFrame(preds).to_csv('results1.csv', index=False)
```
